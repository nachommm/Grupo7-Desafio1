```python
# e4326=CRS('EPSG:4326')
crs = {'init': 'EPSG:4326'}
geometry = [Point(xy) for xy in zip(coordenadas_existentes['lon'], coordenadas_existentes['lat'])]
geometry[:3]
```

```python
geo_coordenadas = gpd.GeoDataFrame(coordenadas_existentes,
                                   geometry = geometry)
geo_coordenadas.head()
```

<h2>Desarrollo un analisis con Bokeh</h2>

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON, get_provider, Vendors 

# p = figure(x_range=(-9180000, -9745000), y_range=(5130000, 5160000),
#            x_axis_type="mercator", y_axis_type="mercator")
#
```

```python
p = figure(x_axis_type="mercator", y_axis_type="mercator")

tile_provider = get_provider(Vendors.CARTODBPOSITRON)

k = 6378137
p.add_tile(tile_provider)

p.circle(x = geo_coordenadas['lon'] * (k * np.pi/180.0),
         y = np.log(np.tan((90 + geo_coordenadas['lat']) * np.pi/360.0)) * k,
         #size=Merged['circle_sizes'],
         line_color="#FF0000", 
         fill_color="#FF0000",
         fill_alpha=0.05)


output_notebook()
show(p)
```
